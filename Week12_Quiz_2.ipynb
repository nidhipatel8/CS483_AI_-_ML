```python
import numpy as np

np.random.seed(100)

dataset=np.random.randint(0,100,size=500)

print(dataset)

print("mean of the original sample",np.mean(dataset))
```

```python

file=np.random.choice(dataset,size=5)
mean_file=[]
dataset2=[]

mean_file
for _ in range(10000):
  file1=np.random.choice(dataset,size=5)
  dataset2.append(file1)
  mean_file.append(file1.mean())

print("list of the mean of the sample is ", mean_file)

print("mean of the sample is given as",np.mean(mean_file))

print("dataset2 is given as ",dataset2)

```
