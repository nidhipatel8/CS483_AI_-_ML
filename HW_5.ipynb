```python
import numpy as np
```

```python
data = np.array([[1,0,0,7,'Apple'],
            [0,1,0,20,'Watermelon'], 
            [1,0,0,1,'Cherry'],
            [0,1,0,7.5,'Apple'],
            [1,0,0,1,'Strawvberry'],
            [1,0,0,0.8,'Cherry']])
```

```python
print(data)
```

```python
X = data[:,:4]
y = data[:,4]
```

```python
print(X)
print(y)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(bootstrap='true',max_features='sqrt',max_depth=4)
model.fit(X,y)
```

```python
test = np.array([[1,0,0,10]])
print(test)
```

```python
model.predict(test)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
estimator = model.estimators_[5]
plt.figure(figsize=(25,10))
a = plot_tree(estimator,
              feature_names = ['Red', 'Green','Blue','Size'],
                class_names = ['Apple','Watermelon','Cherry','Strawberry'],
              filled=True,
              rounded=True,
              fontsize=14)
```
